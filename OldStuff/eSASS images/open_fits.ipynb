```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.tight_layout()
with fits.open("EventFiles\image_6642.fits") as hdul:
    # hdul.info()
    image = hdul[0].data
    plt.figure(dpi=150)
    plt.imshow(np.log(image+1))
    ax = plt.gca()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    plt.savefig('eFEDS_6642.png')
    plt.show()
    # print(hdul[0].header['CDELT2'])

print(np.max(image))
```
